```python
count = 0 
tables = []
for table in soup.find_all('table'):
    tables.append(table.find_all('tr'))
print(tables)
```

```python
flag = 0 
game_titles = []
revenue_values = []
initial_release = []
publishers = []
Genres = []
for row in tables[1]:
    table_division = row.find_all('td')
    if table_division:
        if table_division[0].find('i'):
            game = table_division[0].find('i').string
        else:
            game = table_division[0].find('a').string
        if game == None:
            continue
        revenue = table_division[1].string.rstrip("\n")
        release = table_division[2].string.rstrip("\n")
        if table_division[3].find('a'):
            publisher = table_division[3].find('a').string.rstrip("\n")
        else:
            publisher = table_division[3].string.rstrip("\n")
        print(table_division[4])
        if table_division[4].find('a'):
            genre = table_division[4].find('a').string.rstrip("\n")
        else:
            genre = table_division[4].string.rstrip("\n")
        
        game_titles.append(game)
        revenue_values.append(revenue)
        initial_release.append(release)
        publishers.append(publisher)
        Genres.append(genre)
 
```

```python
gaming_dictionary = {"Game": game_titles, "Revenue": revenue_values, "Initial Release": initial_release, "Publisher(s)": publishers, "Genre(s)": Genres}
```
